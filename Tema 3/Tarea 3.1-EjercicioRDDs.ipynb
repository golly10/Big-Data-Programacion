Crear un array de numeros.

```python
numeros=[2,3,12,56,43,2,23,-4]
numeros
```

Cargar y paralelizar numeros con tres particiones

```python
sc = SparkContext.getOrCreate()

rdd_prl = sc.parallelize(numeros, numSlices=3)
```

Mostrar la lista de números como colección

```python
rdd_prl.collect()
```

Mostrar el número máximo

```python
rdd_prl.max()
```

Mostrar el número mínimo

```python
rdd_prl.min()
```

Acción suma, realizar la reducción a suma

```python
from operator import add

# Inicializamos el acumulador a 0
accum = sc.accumulator(0)

# Recorremos cada elemento de RDD y sumamos cada elemento al acumulador
rdd_prl.foreach(lambda x: accum.add(x))

# Imprimimos el resultado del acumulador
print(accum.value)
```

Cargar  el contenido del fichero  libro.txt  en un rdd. Implementar un contar  palabras y mostrar las todas las  ocurrencias

```python
# Leemos el fichero
rdd_file = sc.textFile("libro.txt")

# Mostramos el resultado
rdd_file.collect()
```

Realizar  la división por palabras asignándole un mapa con valor 1 a cada palabra, mostrando el resultado

```python
words = rdd_file.flatMap(lambda line: line.split())

words_ini = words.map(lambda word: (word, 1))

words_ini.collect()
```

Realizar la reducción por clave, donde cada palabra sea la clave, y el valor la suma de los valores encontrados en cada palabra. Es decir, contar palabras. Mostrar los datos  obtenidos ordenados de forma descendente  por clave.

```python
words_count = words_ini.reduceByKey(lambda wrd1, wrd2: wrd1 + wrd2)

words_sort_key = words_count.sortBy(lambda x: x[1], False)

words_sort_key.collect()
```

Sobre ese conjunto ya reducido y ordenado, obtener los primeros 50 elementos.

```python
words_sort_key.take(50)
```

Obtener el filtrado sobre el original de las palabras de más de 6 letras.

```python
words_more_6 = words.filter(lambda x: len(x) > 6)

words_more_6.collect()
```

Obtener una muestra de ejemplo sobre ese conjunto de palabras  de más de 6 letras, donde no aparezcan repetidas, y una muestra del 25%. Mostrar la muestra obtenida.

```python
words_more_6.filter(lambda x: len(x) > 6).distinct().take(int(len(words.collect())/4))
```
