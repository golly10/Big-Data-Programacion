Crear un array de numeros.

```python
numeros=[2,3,12,56,43,2,23,-4]
numeros
```

Cargar y paralelizar numeros con tres particiones

```python
sc = SparkContext.getOrCreate()

rdd_prl = sc.parallelize(numeros, numSlices=3)
```

Mostrar la lista de números como colección

```python
rdd_prl.collect()
```

Mostrar el número máximo

```python
rdd_prl.max()
```

Mostrar el número mínimo

```python
rdd_prl.min()
```

Acción suma, realizar la reducción a suma

```python
from operator import add

final_sum = 0

for num in rdd_prl.collect():
    final_sum = add(final_sum, num)

print(final_sum)
```
