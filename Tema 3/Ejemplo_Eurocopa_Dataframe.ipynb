```python
from pyspark.sql import SparkSession
spark = SparkSession.builder\
.master("local")\
.appName("Colab")\
.config('spark.ui.port', '4050')\
.getOrCreate()
```

```python
players=spark.read.json("eurocopa.json")
```

```python
players.printSchema()
```

```python
players.count()
```

```python
mayores30=players.filter("age>30")
seleccionados=players.filter("age>30" and "age<35").select("short_name", "long_name","age", "skill_ball_control")
seleccionados.show()
#seleccionados.sort("skill_ball_control",ascending=False).show()
```

```python
class Jugador:
    def __init__(self, nombreCorto, nombreLargo, edad, controlBalon):
        self.nombreCorto = nombreCorto
        self.nombreLargo = nombreLargo
        self.edad = edad
        self.controlBalon = controlBalon
```

```python
datos=seleccionados.toDF("nombreCorto", "nombreLargo", "edad", "controlBalon")
```

```python
valores=datos.collect()
```

```python
for row in valores:
    # while looping through each
    print(row["nombreCorto"],row["edad"])
```
